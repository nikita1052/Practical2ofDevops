```python
def reverse_number_math(n):
    # Determine if the number is negative
    sign = -1 if n < 0 else 1
    n = abs(n)

    reversed_num = 0
    while n > 0:
        remainder = n % 10
        reversed_num = (reversed_num * 10) + remainder
        n = n // 10

    return sign * reversed_num

# Test cases
print("Reverse of number 12345 ",reverse_number_math(12345))
print("Reverse of number -987 ",reverse_number_math(-987))
print("Reverse of number 1200 ",reverse_number_math(1200))
```
