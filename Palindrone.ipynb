```python
def is_palindrome_math(n):

    if n < 0 or (n % 10 == 0 and n != 0):
        return False

    original = n
    reversed_num = 0

    while n > 0:
        remainder = n % 10
        reversed_num = (reversed_num * 10) + remainder
        n = n // 10

    return original == reversed_num

print("12321 is palindrone ",is_palindrome_math(12321))
print("123 is palindrone ", is_palindrome_math(123))
```
